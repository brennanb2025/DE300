## Section 3.2: analyze integer/continuous data plots

There are many outliers in the integer/continous data. I will display the number of outliers in a bar chart. Looking at the box plots, it seems like the values of 0 in features 'trestbps', 'chol', and 'tpeakbpd' are erraneous data, as it is unreasonable for these columns to contain 0 as data.
- For trestbps (resting blood pressure in mm Hg on admission to the hospital), it is unreasonable for a person's blood pressure to be 0.
- For trestbpd (resting blood pressure), it is unreasonable for a person's resting blood pressure to be 0.
- For chol (serum cholestoral in mg/dl), it is unreasonable for a person's cholestoral to be 0.

So, I will remove all 0 values in these columns.

Next, I also believe the three outliers in the column 'cyr' are incorrect data. It is unreasonable for the cyr (year of cardiac cath (sp?)) to not be around the other points, all around 81-86. The value of 1 could be corrupted data, or it could be a typo for 81, and the values of 16 and 26 are likely typos that should be 87. I believe the value of 1 is simply bad data, as all of the data with cyr=81 are grouped at the top of the dataset, yet this one is right in the middle.
So, I will remove the value of 1 but replace the values of 16 and 26 with 86.

Finally, I believe the remaining data is true, valid data. However, I will remove outliers that lie outside the range of 2* to make the columns more normally distributed.

```python
# Impute data: check outliers in continuous and integer columns (mean +- 2/3 standard deviations)

continuous_and_integer_df = df[continuous + integer]

num_outliers_per_col_2_std = {}
num_outliers_per_col_3_std = {}
for col in continuous_and_integer_df.columns:
    mean = continuous_and_integer_df[col].mean()
    std = continuous_and_integer_df[col].std()
    num_outliers_per_col_2_std[col] = continuous_and_integer_df[(continuous_and_integer_df[col] < mean - 2 * std) | (continuous_and_integer_df[col] > mean + 2 * std)][col].count()
    num_outliers_per_col_3_std[col] = continuous_and_integer_df[(continuous_and_integer_df[col] < mean - 3 * std) | (continuous_and_integer_df[col] > mean + 3 * std)][col].count()

sorted_outliers_2_count = dict(sorted(num_outliers_per_col_2_std.items(), key=lambda item: item[1], reverse=True))
sorted_outliers_3_count = dict(sorted(num_outliers_per_col_3_std.items(), key=lambda item: item[1], reverse=True))

# Bar chart
plt.figure(figsize=(10, 6))
plt.bar(sorted_outliers_2_count.keys(), sorted_outliers_2_count.values(), color='skyblue')
plt.title('Number of Outliers per Column (2)')
plt.xlabel('Columns')
plt.ylabel('Number of Outliers')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# bar chart
plt.figure(figsize=(10, 6))
plt.bar(sorted_outliers_3_count.keys(), sorted_outliers_3_count.values(), color='skyblue')
plt.title('Number of Outliers per Column (3)')
plt.xlabel('Columns')
plt.ylabel('Number of Outliers')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Impute data: remove outliers (continued)

# After looking at the outlier data in each column, I believe the outliers
# that are 0 in trestbps, chol, and trestbpd are bad data.
# for these columns, it is unreasonable for the data to be 0.
# For trestbps (resting blood pressure in mm Hg on admission to the hospital),
#   it is unreasonable for a person's blood pressure to be 0.
# For trestbpd (resting blood pressure), it is unreasonable for a person's
#   resting blood pressure to be 0.
# For chol (serum cholestoral in mg/dl), it is unreasonable for a person's
#   cholestoral to be 0.
# So, I will remove all 0 values in these columns.

zero_outlier_cols = ['trestbps', 'trestbpd', 'chol']
for col in zero_outlier_cols:
    df = df[df[col] != 0]


# I also think the three outliers in the column 'cyr' are incorrect data.
# It is unreasonable for the cyr (year of cardiac cath (sp?)) to not be around
#   the other points, all around 81-86.
# The value of 1 could be corrupted data, or it could be a typo for 81
#   and the values of 16 and 26 are likely typos that should be 87.
#   I believe the value of 1 is simply bad data, as all of the data with cyr=81
#   are grouped at the top of the dataset, yet this one is right in the middle.
#   So, I will remove this data.

df['cyr'] = df['cyr'].replace(to_replace=1, value=None)
df['cyr'] = df['cyr'].replace(to_replace=16, value=86)
df['cyr'] = df['cyr'].replace(to_replace=26, value=86)

# I believe the remaining data is true, valid data. However, I will remove
# outliers past 2* to make the columns more normally distributed.
for col in df.columns:
    if col in continuous or col in integer:
        mean = df[col].mean()
        std = df[col].std()
        df.loc[(df[col] < mean - 2 * std) | (df[col] > mean + 2 * std), col] = None
```

Let's visualize the continous and integer columns again after removing the outliers and handling the erraneous data.

```python
# Visualize continuous and integer columns again after removing outliers

continuous_and_integer_df = df[continuous + integer]

# Box plots for each column
plt.figure(figsize=(10, 6))
continuous_and_integer_df.boxplot()
plt.title('Box Plot of Continuous and Integer Columns (Null values ignored)')
plt.xticks(rotation=45)
plt.xlabel('Columns')
plt.ylabel('Values')
plt.show()
```

Wow, look at that! The data doesn't seem to include many outliers.

However, we are about to replace null values with the mean value of the feature, for integer and continuous features, or the median value of the feature, for binary and categorical features. For features that had a lot of missing values, adding many new values will shrink the interquartile range, and produce new outliers. This behavior is expected, and replacing those values with the mean is the best solution.

## Section 3.3: replace all null values with reasonable data

We will replace all null values in binary and categorical features with the median value in order to preserve their categorical nature while filling in the missing data with reasonable data.

We will replace all null values in continuous and integer features with the mean value in order to fill in the missing data with reasonable data. (It is ok to add decimal values into the integer dataset because it will not affect the meaning of the data itself - these features are still providing numerical data.)

```python
# Impute data: remove null values

# Replace null values with median for categorical columns
for col in categorical:
    median_val = df[col].median()
    df[col].fillna(median_val, inplace=True)

# Replace null values with median for binary columns
for col in binary:
    median_val = df[col].median()
    df[col].fillna(median_val, inplace=True)

# Replace null values with mean for integer columns
for col in integer:
    mean_val = df[col].mean()
    df[col].fillna(mean_val, inplace=True)

# Replace null values with mean for continuous columns
for col in continuous:
    mean_val = df[col].mean()
    df[col].fillna(mean_val, inplace=True)

null_percentage = (df.isnull().sum() / len(df)) * 100
print("Number of columns that contain any null data after imputation:", len(null_percentage[null_percentage != 0]))
```

Let's visualize the continuous and integer columns again!

```python
# Visualize continuous and integer columns again.
# This box plot will show more outliers for certain features that had a lot of missing values
# because the null values and outliers that we imputed became the mean,
# shrinking the interquartile range.
# This behavior is expected, and replacing those values with the mean
# is the best solution.

continuous_and_integer_df = df[continuous + integer]

# Box plots for each column
plt.figure(figsize=(10, 6))
continuous_and_integer_df.boxplot()
plt.title('Box Plot of Continuous and Integer Columns (after imputation)')
plt.xticks(rotation=45)
plt.xlabel('Columns')
plt.ylabel('Values')
plt.show()
```

## Section 4.2: compute and analyze various statistical measures

```python
# Compute and analyze the various statistical measures

# compute mean, median, and mode of each feature
print("Mean, median, and mode of each feature. '<lambda>' is the mode.")
df.agg(['mean', 'median', lambda x: x.mode().iloc[0]], axis=0)
```

```python
# kurtosis

from scipy.stats import kurtosis

print("Kurtosis values:")

# Values larger than 5 considered high??
threshold_high = 2
threshold_low = -2

def print_colored(value):
    if value > threshold_high or value < threshold_low:
        print(f"\033[91m{value}\033[0m")  # red color
    else:
        print(value)

for col in continuous:
    kurtosis_value = kurtosis(df[col])
    print(f"Continuous column '{col}': ", end='')
    print_colored(kurtosis_value)

for col in integer:
    kurtosis_value = kurtosis(df[col])
    print(f"Integer column '{col}': ", end='')
    print_colored(kurtosis_value)

for col in categorical:
    kurtosis_value = kurtosis(df[col])
    print(f"Categorical column '{col}': ", end='')
    print_colored(kurtosis_value)
```

Abnormally distributed features: ca and proto

```python
# skewness

def print_colored(value):
    if value > threshold_high or value < threshold_low:
        print(f"\033[91m{value}\033[0m")  # red color
    else:
        print(value)

print("Skewness values:")

# Values larger than 1 or less than -1 considered abnormal.
threshold_high = 1
threshold_low = -1

for col in continuous:
    skewness_value = df[col].skew()
    print(f"Continuous column '{col}': ", end='')
    print_colored(skewness_value)

for col in integer:
    skewness_value = df[col].skew()
    print(f"Integer column '{col}': ", end='')
    print_colored(skewness_value)

for col in categorical:
    skewness_value = df[col].skew()
    print(f"Categorical column '{col}': ", end='')
    print_colored(skewness_value)
```

Abnormally distributed features: ca, proto, and thal

## Section 4.3: visualize the continuous and integer features' distributions with Q-Q plots

Let's now check the Q-Q plots.

# Section 5: apply transformations to make non-normal features normal

**Applying transformations**

The features that are clearly non-normal are:
- ca (kurtosis and skewness flagged it as non-normal: skewed left. Skewness: 1.85, kurtosis: 4.75). ca is continuous.
- proto (kurtosis and skewness flagged it as non-normal: skewed left. Skewness: 2.88, kurtosis: 6.31). proto is categorical.
- thal (skewness flagged it as non-normal: skewed right. Skewness: 1.34, kurtosis: -0.10). thal is categorical.

The features that may be non-normal are:
- rldv5e (Q-Q plot flagged it as potentially non-normal, but it has a skewness of 0.95 and a kurtosis of -0.42). rldv5e is integer.
- oldpeak (Q-Q plot flagged it as potentially non-normal, but it has a skewness of 0.90 and kurtosis of -0.20). oldpeak is integer.
- cigs (Q-Q plot flagged it as potentially non-normal, but it has a skewness of 0.65 and kurtosis of 0.78). cigs is integer.
- years (Q-Q plot flagged it as potentially non-normal, but it has a skewness of 0.30 and kurtosis of -0.06). years is integer.

Now, which of these should we transform?
- ca: yes, this is clearly non-normal.
- proto: categorical. This may just have to do with the categorical encoding.
- thal: categorical. This may just have to do with the categorical encoding.
- rldv5e: it seems to be skewed left. We can attempt to make this column more normal.
- oldpeak: it seems to be skewed left. We can attempt to make this column more normal.
- cigs and years: these two columns are related. cigs refers to the number of cigarettes the person has per day, and years refers to the number of years they were as a smoker. There are a lot of 0 values here, which correspond to non-smokers. These non-smokers are a valuable part of the dataset, and the difference between smoking and not smoking is a valuable differentiator. I think we can attempt to make these columns more normal.

Let's first look at 'ca'.

```python
from scipy.stats import kurtosis

# Function to plot log, sqrt, and normal data.
def q_q_plot_transforms(colName):

    log_transformed_data = np.log(df[colName]+0.000001) # add a small value to avoid log(0).

    sqrt_transformed_data = np.sqrt(df[colName])

    num_plots = 3
    num_cols = 3
    num_rows = 1

    fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 5))
    axes = axes.flatten()

    # Q-Q plot
    ax = axes[0]
    sm.qqplot(df['oldpeak'], line='s', ax=ax)
    ax.set_title(f"Q-Q Plot for oldpeak (unedited data)")
    ax.set_xlabel("Theoretical Quantiles")
    ax.set_ylabel("Ordered Values")
    ax.grid(True)

    ax = axes[1]
    sm.qqplot(log_transformed_data, line='s', ax=ax)
    ax.set_title(f"Q-Q Plot for {colName} (logarithmic transform)")
    ax.set_xlabel("Theoretical Quantiles")
    ax.set_ylabel("Ordered Values")
    ax.grid(True)

    ax = axes[2]
    sm.qqplot(sqrt_transformed_data, line='s', ax=ax)
    ax.set_title(f"Q-Q Plot for {colName} (square root transform)")
    ax.set_xlabel("Theoretical Quantiles")
    ax.set_ylabel("Ordered Values")
    ax.grid(True)


    # for j in range(num_plots, len(axes)):
    #     axes[j].axis('off')

    plt.tight_layout()
    plt.show()

    print(f"Column '{colName}' has a skewness of {df[colName].skew()} and a kurtosis of {kurtosis(df[colName])}.")
    print(f"The kurtosis value for the logarithmic transformed '{colName}' column is:", log_transformed_data.skew())
    print(f"The skewness value for the logarithmic transformed '{colName}' column is:", kurtosis(log_transformed_data))

    print(f"The kurtosis value for the square root transformed '{colName}' column is:", sqrt_transformed_data.skew())
    print(f"The skewness value for the square root transformed '{colName}' column is:", kurtosis(sqrt_transformed_data))
```

```python
# Apply transformations to make these features normally distributed.
# Starting off with ca:
q_q_plot_transforms('ca')
```

**Out of these transforms, the square root transform made the data exhibit the most normal distribution. Let's apply the transform to the column.**

```python
df['ca'] = np.sqrt(df['ca'])
```

Next, let's look at 'rldv5e'

```python
# Apply transformations to make these features normally distributed.
# Next, rldv5e
q_q_plot_transforms('rldv5e')
```

**Both of these transforms made the skewness or kurtosis values worse. We should keep rldv5e as it is.**

Next, let's look at oldpeak.

```python
# Apply transformations to make these features normally distributed.
# Next, oldpeak
q_q_plot_transforms('oldpeak')
```

**Both of these transforms made the skewness or kurtosis values worse. We should keep oldpeak as it is.**

Next, let's look at cigs and years.

```python
# Apply transformations to make these features normally distributed.
# Next, cigs and years
print("cigs:")
q_q_plot_transforms('cigs')

print("years:")
q_q_plot_transforms('years')
```

**Neither of these transforms significantly improved the skewness or kurtosis values for cigs or years. We should keep the cigs and years columns as they are.**

**So, that's it for the integer and continuous features. We just had to apply a square root transform the 'ca' feature. Now let's look at the two non-normal categorical datasets: proto and thal.**

Really, it doesn't make sense to apply a transform to proto and thal to make them normally distributed. They both have integer values that correspond to a certain meaning (for example, for thal, "3 = normal; 6 = fixed defect; 7 = reversable defect"), which are highly arbitrary encodings. The encodings could just as well be strings. Therefore, while the bar charts displayed before do represent the data, because the column values do not hold any meaningful numerical value when compared to each other (as they are simply encodings), we could very well swap two encodings to produce a different distribution. In short, the bars could be arranged in any order by changing the encoding process and it would appear as a different distribution but still accurately represent the data. Therefore, I will not change the encoding of these columns to make them appear normally distributed.


So, we are done! Let's display the cleaned and imputed data one more time.
