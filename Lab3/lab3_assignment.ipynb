```python
#!pip install boto3
#!pip install pandas
#!pip install matplotlib

import boto3
from io import BytesIO
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# you need to change the credentials for yourself

bucket_name = 'de300spring2024'
# object_key = 'brennan_benson/sample_dataset.csv'
object_key = 'brennan_benson/On_Time_Reporting_Carrier_On_Time_Performance_2019_1.csv'
```

```python
# load csv
csv_obj = s3.get_object(Bucket=bucket_name, Key=object_key)
body = csv_obj['Body']
csv_string = body.read().decode('utf-8')
```

```python
# load dataframe
use_cols = [
    "DayofMonth", "DayOfWeek", "DepTime", "CRSDepTime",
    "ArrTime", "CRSArrTime", "Reporting_Airline","Flight_Number_Reporting_Airline",
    "Tail_Number", "ActualElapsedTime", "CRSElapsedTime", "AirTime", "ArrDelay", "DepDelay", "Origin", "Dest",
    "Distance", "TaxiIn", "TaxiOut", "Cancelled",
    "CancellationCode", "Diverted", "CarrierDelay", "WeatherDelay", "NASDelay", "SecurityDelay",
    "LateAircraftDelay", "OriginStateName", "DestStateName", # my stuff
]
jan = pd.read_csv(
    BytesIO(csv_string.encode()),usecols=use_cols
)
# randomly sample 100k observations and perform analysis on this subset
jan = jan.sample(n=100000,random_state=1)
jan.rename(
    columns={
        "Reporting_Airline":"UniqueCarrier"
    },
    inplace=True
)
```

```python
# Calculate average delay components
avg_delay_components = jan[['CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay']].mean()

# Plot
plt.figure(figsize=(8, 6))
avg_delay_components.plot(kind='bar', color='teal')
plt.xlabel('Delay Components')
plt.ylabel('Average Delay Time (minutes)')
plt.title('Average Delay Components for Flights')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Explanation of the Average Delay Components for Flights chart:

The plot represents the average contribution from each type of delay to each delay that happened.

Over all of the states and all of the carriers, the largest causes of delay time by far were carrier delay, NAS (National Air System) delay, and late aircraft delay. Security delay contributed almost no major delay time, and weather delay contributed an average of 5 minutes of delay.

```python
avg_weather_delay = jan.groupby('OriginStateName')['WeatherDelay'].mean().reset_index()

# average weather delay, descending order
top_10_states_with_longest_delay = avg_weather_delay.sort_values(by='WeatherDelay', ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x='OriginStateName', y='WeatherDelay', data=top_10_states_with_longest_delay)
plt.xlabel('State')
plt.ylabel('Average Weather Delay (minutes)')
plt.title('Top 10 States with Longest Average Weather Delay')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Explanation of the Top 10 States with Longest Average Weather Delay chart:

The plot represents the average weather delay, for flights that experienced a delay, for different states of origin (the states from which the plane departed). It only displays the 10 states with the largest weather delays.

The data was from January, which is the coldest month for many of these states. According to [this](https://www.thrillist.com/travel/nation/us-states-with-the-worst-winters) article titled "US States with the Worst Winters", the 5 states with the most brutal winters were North Dakota, Minnesota, Michigan, South Dakota, and Wisconsin. 4 of these 5 states are represented on this graph - it is likely that the cold weather and snow in these 5 states caused much airline weather delay. All of the states represented in this top 10 graph have cold winters.

As for Rhode Island, I think its appearance on this graph could be explained by its small sample size of airports and landings, as it is such a small state.

```python
avg_carrier_delay = jan.groupby('UniqueCarrier')['CarrierDelay'].mean().reset_index()

# average carrier delay, descending order
top_10_carriers_with_longest_delay = avg_carrier_delay.sort_values(by='CarrierDelay', ascending=False).head(5)

plt.figure(figsize=(6, 6))
sns.barplot(x='UniqueCarrier', y='CarrierDelay', data=top_10_carriers_with_longest_delay)
plt.xlabel('Carrier')
plt.ylabel('Average Carrier Delay (minutes)')
plt.title('Top 5 Carriers with Longest Average Carrier Delay')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Explanation of the Top 5 Carriers with Longest Average Carrier Delay chart:

The plot represents the average carrier delay, for flights that experienced a delay, for different carriers. It only displays the 5 carriers with the largest carrier delays. This graph therefore shows just the delay caused by the carrier itself - not weather, the national air system, or late aircrafts.

The top 5 airline carriers with the longest average carrier delays were SkyWest Airlines (OO), Mesa Airlines, Inc. (YV), Jetblue Airways Corporation (B6), Delta Air Lines, Inc. (DL), and Hawaiian Airlines, Inc. (HA).

SkyWest and Mesa are regional carriers, while Jetblue and Delta are major carriers, and Hawaiian Airlines only serves flights to and from Hawaii.

This chart could help potential flyers decide what airlines to avoid.
